```python
import numpy as np
import matplotlib.pyplot as plt

# to see the plots inside jupyter notebook
%matplotlib inline
```

#### creating the data to plot

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

#### Visualization

```python
x
```

```python
y
```

#### Functional PLOT

```python
# FUNCTIONAL

# creating the plot
plt.plot(x, y)
# setting x and y label
plt.xlabel('X Label')
plt.ylabel('Y Label')
# setting the title
plt.title('My Plot')

plt.show()
```

#### Subplots

```python
#           (n_rows, n_cols, figure) 
plt.subplot(1,2,1)
# plot refering to fig 1 in subplot
plt.plot(x,y, 'r')
plt.title('First Plot')

plt.subplot(1,2,2)
# plot refering to fig 2 in subplot
plt.plot(x,y,'b')
plt.title('Second Plot')

plt.show()
```

#### OO Plot

```python
# OO

# Create a figure object
fig = plt.figure()
# set axes for the figure obj
axes = fig.add_axes([0.1, 0.1, 1, 1])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('OO Plot')

fig.show()
```

#### OO Subplots

```python
fig = plt.figure()

# axes = [left, bottom, width, height] the values are between (0,1)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Amount')
axes1.plot(x,y, 'g')

axes2 = fig.add_axes([0.63, 0.17, 0.25, 0.25])
axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes2.set_title('Error')
axes2.plot(y,x, 'r')

fig.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)

# for ax in axes:
#     ax.set_xlabel('X Label')
#     ax.set_ylabel('Y Label')
#     ax.plot(x,y)

axes[0].plot(x,y)
axes[1].plot(y,x)

fig.show()
```

### Fig size and DPI

```python
fig = plt.figure(figsize=(5,3), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

fig.show()
# fig.savefig('name_of_fig.jpg')
```

#### Legend

```python
fig = plt.figure(figsize=(7,5))

ax = fig.add_axes([0,0,1,1])
# to see the legend we need to set a label for each axes
ax.plot(x, x**3, label='X Cube')
ax.plot(x, x**2, label='X Squared')
# set the legend (loc it's the location to show, loc=0 the matplotlib found the best place to show)
# ax.legend(loc=10)
# or pass a touple with the X and Y position
ax.legend(loc=(0.5,0.5))

fig.show()
```

### Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# color parameter must be a string color name or rgb hex code #fffff
# linewitdh = lw
# linestyle = ls
# alpha control the transparency of line
# marker, markersize, markerfacecolor, markeredgewidth, markeredgecolor
ax.plot(x,y, color='#303030', linewidth=4, alpha=0.7, linestyle='-', marker='o', markersize=10)

# set x and y limits
ax.set_xlim([0,4])
ax.set_ylim([0,5])

fig.show()
```
