```python
import numpy as np
import matplotlib.pyplot as plt
```

# Data

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

### Exercise 1

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('First Exercise')

fig.show()
```

### Exercise 2

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,0.5,0.2,0.2])

ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Exercise 2')
ax1.plot(x,y, 'r')

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Inside Plot')
ax2.plot(x,y, 'r')

fig.show()
```

### Exercise 3

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

ax1.plot(x,y,'b')
ax1.plot(x,z,'g')
ax2.plot(x,z)

ax2.set_xlim(20,22)
ax2.set_ylim(300,500)
ax2.set_title('zoom')

fig.show()
```

### Exercise 4

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,3))

axes[0].plot(x,y, linestyle='--', linewidth=4)

axes[1].plot(x,z, linewidth=4, color='r')

fig.show()
```
