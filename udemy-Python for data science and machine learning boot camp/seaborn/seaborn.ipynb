```python
import seaborn as sns
import numpy as np

%matplotlib inline
```

### Load Tips Data

```python
tips = sns.load_dataset('tips')
tips.head()
```

### Distribution Plot

```python
# Histogram

# kde=False remove the line
# kde = Kernel Density Estimation
sns.distplot(tips['total_bill'], kde=False, bins=40)
```

### Joint Plot

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

### Pair Plot

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

### Rug plot

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

## Categorical Plots

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
# Count the occurences
sns.countplot(x='day', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='coolwarm')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='coolwarm', hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips) #hue='sex', split=True
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```
